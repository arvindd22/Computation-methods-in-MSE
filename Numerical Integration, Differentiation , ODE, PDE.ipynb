```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return x**(3)/3

# Define the exact derivative of the function
def f_prime(x):
    return x**2

# Define functions for central and forward difference methods
def central_difference(f, x, h):
    return (f(x + h) - f(x - h)) / (2 * h)

def forward_difference(f, x, h):
    return (f(x + h) - f(x)) / h

# Define the true derivative values
x_values = np.linspace(-1, 1, 1000)
true_derivative = f_prime(x_values)

# Arrays to store error values
central_difference_errors = []
forward_difference_errors = []
h_values = []

# Calculate errors for different h values
for i in range(2, 10):
    h = 10**(-i)
    h_values.append(h)

    # Calculate derivatives using both methods
    central_derivatives = [central_difference(f, x, h) for x in x_values]
    forward_derivatives = [forward_difference(f, x, h) for x in x_values]

    # Calculate errors
    central_error = np.abs(central_derivatives - true_derivative)
    forward_error = np.abs(forward_derivatives - true_derivative)

    # Calculate and store maximum errors
    central_difference_errors.append(np.max(central_error))
    forward_difference_errors.append(np.max(forward_error))

# Plotting the errors
plt.figure(figsize=(10, 6))

# Plotting central difference errors
plt.subplot(1, 2, 1)
plt.loglog(h_values, central_difference_errors, label='Central Difference')
plt.title('Central Difference Error')
plt.xlabel('h (log scale)')
plt.ylabel('Error (log scale)')
plt.legend()

# Plotting forward difference errors
plt.subplot(1, 2, 2)
plt.loglog(h_values, forward_difference_errors, label='Forward Difference')
plt.title('Forward Difference Error')
plt.xlabel('h (log scale)')
plt.ylabel('Error (log scale)')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#SOLVE ODE
import numpy as np
import matplotlib.pyplot as plt

# Define the derivative function  dx/dt=-te^(-x)
def derivative(t, x):
    return -t * np.exp(-x)

# Euler method for solving the ODE
def euler_method(f, t0, x0, h, num_steps):
    t = np.zeros(num_steps + 1)
    x = np.zeros(num_steps + 1)

    t[0] = t0
    x[0] = x0

    for i in range(num_steps):
        x[i+1] = x[i] + h * f(t[i], x[i])
        t[i+1] = t[i] + h
        print("t=", t[i],"  x=",x[i])

    return t, x

# Initial conditions and parameters
t0 = 0
x0 = 1
h = 0.1  # Step size
num_steps = 100

# Solve the ODE using Euler method
t, x = euler_method(derivative, t0, x0, h, num_steps)

# Plot the solution
plt.figure(figsize=(8, 6))
plt.plot(t, x, label='Solution x(t)')
plt.xlabel('t')
plt.ylabel('x')
plt.title('Solution of dx/dt = -t * exp(-x) using Euler method')
plt.legend()
plt.grid()
plt.show()
```
