```python
import numpy as np
import matplotlib.pyplot as plt
x=float(input("Enter initial guess:"))
err=[]
for n in range(50):
  x0= x
  x = np.exp(-x**2)
  err.append(x0-x)
  print(x," ",err[n], " ",n)
  if abs(err[n])< 0.01:
    break
print("The solution is:", x)
plt.ylabel("Error")
plt.xlabel("Iterations")
plt.plot(range(n+1), err,"-o")
plt.show()
```

```python
import matplotlib.pyplot as plt
#Define the function and first derivative
def f(x):
  return x**3 - 15*x -4
def df(x):
  return 3*x**2 - 15
def newton(x1):
  err=[]
  for i in range(500):
    x2 =x1- f(x1)/ df(x1)
    err.append(x2-x1)
    if(abs(x2-x1)<=0.00000000001):
      break
    x1=x2
  print("Root at:", x2, "Converged in:",i, "steps")
  print("Value of the function:", f(x2))
  plt.plot(range(i+1),err,"-o")
  plt.show()
#main code
x1= float(input("Enter the value of x1:"))
newton(x1)
```
