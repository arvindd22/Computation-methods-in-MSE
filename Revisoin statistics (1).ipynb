```python
import numpy as np
from scipy.stats import norm
a = 4
b = 6
n = 30
# Calculate the mean and standard deviation of the sample mean (X̄)
sample_mean_mu = (a + b) / 2
sample_mean_sigma = (b - a) / (2 * np.sqrt(n))

# Define the range for which you want to calculate the probability
range_start = 4.8
range_end = 5.2

# Calculate the z-scores for the range
z_start = (range_start - sample_mean_mu) / sample_mean_sigma
z_end = (range_end - sample_mean_mu) / sample_mean_sigma

# Use the cumulative distribution function (CDF) of the standard normal distribution
# to calculate the probability
probability = norm.cdf(z_end) - norm.cdf(z_start)

print("Probability that the sample mean is between {} and {}: {:.4f}".format(range_start, range_end, probability))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
population_variance = 20
sample_size = 10
null_mean = 100
alpha = 0.05

# Calculate the standard error
standard_error = np.sqrt(population_variance / sample_size)

# Create a range of alternative means for the power function
alternative_means = np.linspace(95, 105, 100)  # Adjust the range as needed

# Initialize arrays to store power values for different alternative means
power_right_sided = []
power_left_sided = []
power_two_sided = []

# Calculate power for each alternative mean
for alternative_mean in alternative_means:
    # Calculate the critical values for the null hypothesis
    z_critical_right = norm.ppf(1 - alpha)  # Right-sided test
    z_critical_left = -norm.ppf(1 - alpha)  # Left-sided test
    z_critical_two_sided = norm.ppf(1 - alpha / 2)  # Two-sided test

    # Calculate the Z-scores for the alternative mean
    z_right = (alternative_mean - null_mean) / standard_error
    z_left = (alternative_mean - null_mean) / standard_error

    # Calculate power for each type of test
    power_right = 1 - norm.cdf(z_critical_right - z_right)
    power_left = norm.cdf(z_critical_left - z_left)
    power_two = 1 - norm.cdf(z_critical_two_sided - abs(z_right))

    power_right_sided.append(power_right)
    power_left_sided.append(power_left)
    power_two_sided.append(power_two)

# Plot the power functions
plt.figure(figsize=(10, 6))
plt.plot(alternative_means, power_right_sided, label="Right-Sided Test")
plt.plot(alternative_means, power_left_sided, label="Left-Sided Test")
plt.plot(alternative_means, power_two_sided, label="Two-Sided Test")
plt.axhline(y=alpha, color='r', linestyle='--', label=f'Significance Level (α = {alpha})')
plt.xlabel("Alternative Mean")
plt.ylabel("Power")
plt.title("Power Functions for Different Hypothesis Tests")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Given parameters
α = 0.05
σ_squared = 20
n = 10
μ0 = 100

# Calculate critical value
z_critical = stats.norm.ppf(α)

# Calculate standard error of the mean (SEM)
SEM = np.sqrt(σ_squared / n)

# Specify a range of μ1 values for the power function
μ1_values = np.linspace(90, 100, 100)  # Adjust the range as needed

# Calculate the power for each μ1 value
power_values = 1 - stats.norm.cdf(z_critical - (μ1_values - μ0) / SEM)

# Plot the power function
plt.figure(figsize=(8, 6))
plt.plot(μ1_values, power_values)
plt.xlabel("μ1 (Alternative Mean)")
plt.ylabel("Power")
plt.title("Power Function")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Given parameters
α = 0.05
σ_squared = 20
n = 10
μ0 = 100

# Calculate critical values
z_critical_left = stats.norm.ppf(α / 2)
z_critical_right = stats.norm.ppf(1 - α / 2)

# Calculate standard error of the mean (SEM)
SEM = np.sqrt(σ_squared / n)

# Specify a range of μ1 values for the power function
μ1_values = np.linspace(90, 110, 100)  # Adjust the range as needed

# Calculate the test statistic (z) for the range of μ1 values
z_values = (μ1_values - μ0) / SEM

# Calculate the power of the test for each μ1 value
power_values = 1 - stats.norm.cdf(z_values - z_critical_right) + stats.norm.cdf(z_values - z_critical_left)

# Plot the power function
plt.figure(figsize=(10, 7))
plt.plot(μ1_values, power_values)
plt.xlabel("μ1 (Alternative Mean)")
plt.ylabel("Power")
plt.title("Power Function for Two-Sided Test")
plt.grid(True)
plt.show()
```
