```python
import pandas as pd
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import numpy as np
from google.colab import files
uploaded= files.upload()
strength_df=pd.read_csv('linreg.csv')

X=sm.add_constant(strength_df['Composition'])
Y=strength_df['Strength']

train_X, test_X, train_Y, test_Y =train_test_split( X, Y, train_size=0.6)
strength_lrm= sm.OLS(train_Y, train_X).fit
print(strength_lrm.params)
pred_Y=strength_lrm.predict(test_X)
print("R-Sq = ", np.abs(r2_score(test_Y,pred_Y )))
```
